### **TASK 4 - IMAGE TO PENCIL SKETCH WITH PYTHON**

**Imports**

```python
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Reading image file

```python
img = cv2.imread("/content/bird.jpg")
```

Converting an image from BGR color space to RGB color space using OpenCV and then displaying the image using matplotlib in a Python environment.

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis(False)
plt.show()
```

Converting an image from BGR color space to grayscale using OpenCV and then attempting to display the grayscale image using plt.imshow() from matplotlib

```python
gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gimg)
```

Converting a grayscale image to its inverted form using OpenCV's cv2.bitwise_not() function and then attempts to display the inverted image using plt.imshow() from matplotlib.

```python
invert_img = cv2.bitwise_not(gimg)
plt.imshow(invert_img)
```

Visualizing the inverted grayscale image correctly, using the cmap parameter in plt.imshow() to specify the colormap for grayscale images.

```python
plt.imshow(invert_img, cmap='gray')
plt.axis(False)
plt.show()
```

Applying Gaussian blurring to the inverted image using OpenCV's cv2.GaussianBlur() function and attempts to display the blurred image using plt.imshow() from matplotlib.

```python
blurr_img = cv2.GaussianBlur(invert_img, (111,111), 0)
plt.imshow(blurr_img)
```

 Inverting the blurred image by using cv2.bitwise_not() from OpenCV and then attempts to display the inverted blurred image using plt.imshow() from matplotlib.

```python
invert_blurr_img = cv2.bitwise_not(blurr_img)
plt.imshow(invert_blurr_img)
```

Creating a sketch-like image by dividing the grayscale image (gimg) by the inverted blurred image (invert_blurr_img) using cv2.divide() from OpenCV.

```python
sketch_image = cv2.divide(gimg, invert_blurr_img, scale=256.0)
plt.imshow(sketch_image)
```

Saving the sketch image as a PNG file using OpenCV's cv2.imwrite() function.

```python
cv2.imwrite("sketch.png", sketch_image)
```

Adjusting the size of the figure using plt.figure(figsize=(8, 8)). Then, it converts the sketch image from RGB to BGR color space using cv2.cvtColor() from OpenCV. Finally, it displays the image using plt.imshow() from matplotlib

```python
plt.figure(figsize = (8,8))
image3 = cv2.cvtColor(sketch_image, cv2.COLOR_RGB2BGR)
plt.imshow(image3)
plt.axis(False)
plt.show()
```
